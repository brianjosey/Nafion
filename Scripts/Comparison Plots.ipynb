This notebook is just a quick plotting script that I threw together so that I can compare nSLD profiles of two different models on top of each other.

```python
# Essential libraries

import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import seaborn as sns
```

```python
def load_profile(path: str):
    """
    Takes the specified profile, cleans it, and loads it into a pandas dataframe
    
    Args:
        path (str): filename
        
    Returns:
        profile (df): fitted df containing z, rho, rhoM, and theta
    """
    profile = pd.read_csv(f"{path}.dat",
                         delim_whitespace=True,
                         usecols=["#","z","rho","(1e-6/A2)"])
    profile.columns = ["z", "rho", "rhoM", "theta"]
    return profile
```

```python
# Load profiles
ar_profile = load_profile("Ar_lamellae-profile")
d2o_profile = load_profile("D2O_lamellae-profile")
h2o_profile = load_profile("H2O_lamellae-profile")
```

```python
# Plot reflectivity
fig, ax = plt.subplots(figsize = (16, 10))

sns.lineplot(x=ar_profile['z'], y=ar_profile['rho'], sort=False, linewidth = 2, label="Ar Gas (Dry)")
sns.lineplot(x=d2o_profile['z'], y=d2o_profile['rho'], sort=False, linewidth = 2, label="D2O-92 % RH")
sns.lineplot(x=h2o_profile['z'], y=h2o_profile['rho'], sort=False, linewidth = 2, label="H2O-92 % RH")

plt.title("nSLD Profile-Lamellae Layer", fontsize=16, fontweight='bold')
plt.xlabel("Height Above Silicon/Oxide Interface, z/ Å")
plt.ylabel("nSLD/ 10^-6 Å^-2")

plt.show()
```
